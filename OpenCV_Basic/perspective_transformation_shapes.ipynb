```python
# PERSPECTIVE TRANSFORMATION

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/home/srijanani/GIT/OpenCV/Images/shape_2.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
src_points = np.float32([[1000, 1200], [2500, 750], [1500, 2200], [3250, 1750]])
```

```python
dst_points = np.float32([[0,0], [500, 0], [0, 500], [500, 500]])
```

```python
matrix = cv2.getPerspectiveTransform(src_points, dst_points)
```

```python
transformed_images = cv2.warpPerspective(img, matrix, (500, 500))
```

```python
plt.imshow(transformed_images)
```
