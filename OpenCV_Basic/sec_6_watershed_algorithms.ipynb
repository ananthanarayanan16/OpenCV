```python
# WATERSHED ALGORITHM

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(image, cmap='gray'):

    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(image, cmap='gray')
```

```python
sep_coins = cv2.imread('/home/srijanani/GIT/OpenCV/Images/coin.png')
# sep_coins = cv2.cvtColor(sep_coins, cv2.COLOR_BGR2RGB)
# plt.imshow(sep_coins)

display(sep_coins)
```

```python
# Median blur
# Gray Scale
# Binary Thershold
# Find Contour
```

```python
# APPLYING MEDIAN BLUR

sep_blur = cv2.medianBlur(sep_coins,11)

display(sep_blur)
```

```python
# CONVERTING GRAY SCALE THE IMAGE

gray_sep_coins = cv2.cvtColor(sep_blur, cv2.COLOR_BGR2GRAY)

display(gray_sep_coins)
```

```python
 # APPLYING BINARY THRESHOLD

res, sep_threshold = cv2.threshold(gray_sep_coins, 190, 255, cv2.THRESH_BINARY_INV)

display(sep_threshold)
```

```python
contours, hierarchy = cv2.findContours(sep_threshold.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for i in range(len(contours)):

    if hierarchy[0][i][3] == -1:

        cv2.drawContours(sep_coins, contours, i, (255, 0 , 0), 5)
```
