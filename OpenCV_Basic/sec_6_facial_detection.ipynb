```python
face_cascade = cv2.CascadeClassifier('/home/srijanani/GIT/OpenCV/Images/haarcascade.xml')
```

```python
def detect_face(img):

    face_image = img.copy()

    face_rects = face_cascade.detectMultiScale(face_image) #, scaleFactor = 1.1, minNeighbors = 5)

    for (x, y, w, h) in face_rects:

        cv2.rectangle(face_image, (x, y), (x+w, y+h), (255, 255, 255), 7)

    return face_image
```
