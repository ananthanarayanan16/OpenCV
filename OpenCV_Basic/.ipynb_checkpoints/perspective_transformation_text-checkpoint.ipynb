```python
# PERSPECTIVE FUNCTION

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/home/srijanani/GIT/OpenCV/Images/text_3.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


plt.imshow(img)
```

```python
src_points = np.float32([[1750 ,750], [2750, 750], [2000, 1750], [3500, 1750]]) 
```

```python
dst_points = np.float32([[0, 0], [500, 0], [0, 500], [500, 500]])
```

```python
matrix = cv2.getPerspectiveTransform(src_points, dst_points)
```

```python
trans_img = cv2.warpPerspective(img, matrix, (500, 500))
```

```python
plt.imshow(trans_img)
```
