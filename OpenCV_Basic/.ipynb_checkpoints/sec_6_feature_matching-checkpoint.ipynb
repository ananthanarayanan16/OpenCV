```python
# FEATURE MATCHING 

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap='gray'):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
reeses = cv2.imread('/home/srijanani/GIT/OpenCV/Images/reeses.png')
reeses = cv2.cvtColor(reeses, cv2.COLOR_BGR2GRAY)

# plt.imshow(reeses, cmap='gray')
display(reeses)
```

```python
cereals = cv2.imread('/home/srijanani/GIT/OpenCV/Images/cereals.jpg', 0)

display(cereal)
# plt.imshow(cereals, cmap='gray')
```

```python
orb = cv2.ORB_create()
```

```python
kp1, des1 = orb.detectAndCompute(reeses, None)

kp2, des2 = orb.detectAndCompute(cereals, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
```

```python
matches = bf.match(des1, des2)
```

```python
single_matches = matches[0]
```

```python
single_matches
# matches
```

```python
single_matches.distance
```

```python
len(matches)
```

```python
matches = sorted(matches, key=lambda x:x.distance)
```

```python
reeses_matches = cv2.drawMatches(reeses, kp1, cereals, kp2, matches[:6], None, flags=2)
```
