```python
# CONTOUR DETECTION

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/home/srijanani/GIT/OpenCV/Images/circle.png', 0)

# img = cv2.GaussianBlur(image, (5,5), 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# CREATING BLANK IMAGE

external_contours = np.zeros_like(img)
```

```python
# img.shape
```

```python
external_contours.shape
```

```python
# plt.imshow(external_contours)
```

```python
# list(range(len(contours)))
```

```python
# DRAW THE EXTERNAL CONTOURS

for i in range(len(contours)):

    # EXTERNAL CONTOUR
    if hierarchy[0][i][3] == -1:

        cv2.drawContours(external_contours, contours, i, 255, -1)
```

```python
# contours
```

```python
plt.imshow(external_contours, cmap='gray')
```

```python
# DRAW INTERNALL CONTOURS

internal_contours = np.zeros_like(img)


for i in range(len(contours)):

    if hierarchy[0][i][3] == 99:

        cv2.drawContours(internal_contours, contours, i, 255, -1)
```

```python
plt.imshow(internal_contours, cmap='gray')
```
