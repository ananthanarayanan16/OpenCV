```python
import cv2
import numpy as np

# Load an image
image = cv2.imread('/home/srijanani/GIT/OpenCV/Data/home.jpeg')
if image is None:
    raise ValueError("Image not found. Please check the file path.")

# Create a window
cv2.namedWindow('Zoom')

# Function to zoom the image
def zoom(val):
    scale = val / 100  # Convert trackbar value to scale factor
    if scale == 0:  # Avoid scale = 0
        scale = 0.1
    height, width = image.shape[:2]
    
    # Rescale the image
    zoomed = cv2.resize(image, None, fx=scale, fy=scale, interpolation=cv2.INTER_LINEAR)

    # Center crop if zoomed in
    if scale > 1.0:
        center_x, center_y = zoomed.shape[1] // 2, zoomed.shape[0] // 2
        crop_x, crop_y = width // 2, height // 2
        zoomed = zoomed[center_y - crop_y:center_y + crop_y, center_x - crop_x:center_x + crop_x]
    # Pad if zoomed out
    else:
        pad_x = (width - zoomed.shape[1]) // 2
        pad_y = (height - zoomed.shape[0]) // 2
        zoomed = cv2.copyMakeBorder(zoomed, pad_y, pad_y, pad_x, pad_x, cv2.BORDER_CONSTANT)

    # Show the zoomed image
    cv2.imshow('Zoom', zoomed)

# Create a trackbar to control zoom
cv2.createTrackbar('Scale (%)', 'Zoom', 100, 200, zoom)

# Show the original image initially
cv2.imshow('Zoom', image)

# Keep the window open until 'q' is pressed
while True:
    key = cv2.waitKey(1)
    if key == ord('q'):  # Exit on pressing 'q'
        break

cv2.destroyAllWindows()
```
